```python

class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade

    def display(self):
        print(f"Name: {self.name}, Grade: {self.grade}")

    def age(self,ageofstudent):
      self.age=ageofstudent
      print(f"age:{self.age}")

# Create an object
student1 = Student("Niki", "A",19)
student1.display()
student2 = Student("Nik", "B",20)
student2.display()
```

```python

class customer:
    def __init__(self, name, gender,income):
        self.name=name
        self.gender=gender
        self.income=income

    def display(self):
        print(f"name: {self.name}, gender: {self.gender},income:{self.income}")


# Create an object
customer1 = customer("Niki", "F",50000)
customer1.display()
customer2 = customer("Nike", "M",20000)
customer2.display()
customer3 = customer("Mini", "F",95000)
customer3.display()
customer4 = customer("Paru", "F",95000)
customer4.display()
customer5 = customer("Meenu", "F",89000)
customer5.display()
```
